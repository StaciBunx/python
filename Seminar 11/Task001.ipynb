## Дана функция f(x) = (sin(x)/2 + cos(x)/2)^2 * sin(x)

```python
from sympy import *
x = Symbol('x')
f = trigsimp(((sin(x)/2)+(cos(x)/2))**2*sin(x))
f
```

**1. Определить корни**

```python
res = solveset(f,x)
res
```

**2. Найти интервалы, на которых функция возрастает**

```python
print('Интервал от -pi/2 до 2*pi')
coord_list = sorted(solveset(diff(f, x), x, domain=Interval(-pi/2, 2*pi)))
for i in range(1, len(coord_list)):
    if f.subs(x, coord_list[i]) > f.subs(x, coord_list[i] + 0.0000001):
        print(f'Функция возрастает на отрезке: [{simplify(coord_list[i-1])}; {simplify(coord_list[i])}]\n')
        plot(f, (x, coord_list[i-1], coord_list[i])) 
    else:
        continue
```

**3. Найти интервалы, на которых функция убывает**

```python
print('В интервале -pi/2 до 2*pi\n')
for i in range(0, len(coord_list)):
    if f.subs(x, coord_list[i]) < f.subs(x, coord_list[i] + 0.0000001):
        print(f'Функция убывает на отрезке: [{simplify(coord_list[i-1])}; {simplify(coord_list[i])}]\n')
        plot(f, (x, coord_list[i-1], coord_list[i])) 
    else:
        continue
```

**4. Построить график**


```python
print('В интервале -pi/2 до 2*pi\n')
chart = plot(f,(x,-2*pi,2*pi)) 
```

**5.Вычислить вершину**

```python
extr = solveset(diff(f, x), x, domain=Interval(0, 2*pi))
extr = list(extr)
y_arr = []
x_arr = []
for i in range(0, len(extr)):
    x_arr.append(simplify(extr[i]))
    y_arr.append(f.subs(x, simplify(extr[i])))
y_max_index = y_arr.index(max(y_arr))
print(f'Вершина в интервале [0, 2*pi]:\n')
print(f'x-координата: {x_arr[y_max_index]}\n')
print(f'y-координата: {y_arr[y_max_index]}\n')
```

**6. Определить промежутки, на котором f > 0**

```python
print('В интервале [0, 2*pi] f>0 при х:\n')
solve_univariate_inequality(f > 0, x, relational=False, domain=Interval(0, 2*pi))
```

**7. Определить промежутки, на котором f < 0**

```python
print('В интервале [0, 2*pi] f<0 при х:\n')
solve_univariate_inequality(f < 0, x, relational=False, domain=Interval(0, 2*pi))
```
